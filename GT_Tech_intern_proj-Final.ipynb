## Bag_of_Words representation of the corpus

```python
from gensim.corpora import Dictionary
```

```python
# Creating the dictionary and the corpus

def corp_dict(docs,filt_val,filt_extreme=False):
    """
    This function will take tokenized document list as input, initialize a dictionary with that and create a bag of words
    corpus.It is also filter the words basis value given in filt_val(no_below and no_above) when filter extreme is true.
    
    docs = list of tokenized documents, Ex. [fencing, riat, market, kanyikela, ward, info,..]
    filt_val = list of tuple, tuple containing (no_below,no_above) - ex. [(4,2),(4,0.5)].
                If the value passed is integer, words having repeatation below that particular value will be filtered out
                If given float, percentage availability of the word in the dictionary with respect to its frequency will be
                considered.
                filt_val will only be considered if filt_extreme is TRUE
    filt_extreme = bool, True or False   
    
    Return:
    list of tuples consisting of dictionary and corpus - Ex. [(dictionary,corpus)]
    
    dictionary - iterable dictionary of gensim class. to print the dictionary loop over dictionary.items() and print the content.
    
    corpus - bag of words corpus, list of tuples. length of corpus is length of the document list or no. of documents
             the tuple consists of word id and frequecy of the word in the respective document.
             [(0, 1), (1, 2), (2, 2), (3, 1), (4, 1), (5, 1), (6, 2), (7, 1), (8, 1), (9, 1), (10, 1), (11, 1), (12, 2), (13, 1), 
             (14, 1),(15, 1), (16, 1), (17, 3), (18, 2), (19, 1), (20, 1)]
             
    
    """
    
    # Initializing Dictionary class instance with the document in discussion# Bag-of-words representation of the documents.
    lst=[]
     
    if filt_extreme:
        c=0
        for tupl in filt_val:
            c+=1
            dictionary = Dictionary(docs)
            dictionary.filter_extremes(no_below=tupl[0], no_above=tupl[1])
            dictionary.compactify()
            corpus_bow = [dictionary.doc2bow(doc) for doc in docs]
            lst.append((dictionary,corpus_bow))
            print("For Filter value set  {}".format(c))
            print('Number of unique tokens: %d' % len(dictionary))
            print('Number of documents: %d' % len(corpus_bow),'\n')
    else:
        dictionary = Dictionary(docs)
        corpus_bow = [dictionary.doc2bow(doc) for doc in docs]
        lst.append((dictionary,corpus_bow))
        
        print('Number of unique tokens: %d' % len(dictionary))
        print('Number of documents: %d' % len(corpus_bow),'\n')
    
    return lst
        
    
```

```python
corp_dict_lst=corp_dict(docs,[(4,0.5),(2,0.2),(1,0.2)],True)
```

## Creation of the Tfidf Model

```python
from gensim.models import TfidfModel
```

```python
def tfidf_vec(corpus):
    model = TfidfModel(corpus)  # fit mode
    vector_tfidf = model[corpus]
    return vector_tfidf
```

## Latent Dirichlet Allocation using Bag of Words

```python
from gensim.models import LdaModel
```

```python
def lda_app(corpus_lst,topic_lst,passes,chunksize,iterations,eval_every=None):
    ''' 
    This function is a wrapper function and providing data to LdaModel function
    
    corpus_lst - list of tuples consisting of dictionary and bag of words corpus. Ex.[(dictionary,corpus_bow),
    (dictionary,corpus_bow),...]
    
    topic_lst - list of int â The number of requested latent topics to be extracted from the training corpus. For each integer, 
    which is the no.of topics, all the corpus available in the corpus list are run.

    passes (int) â Number of passes through the corpus during training. (may arneg between 10-50 ro more than that.)
    
    chunksize (int) â Number of documents to be used in each training chunk. (Batch size for processing)
    
    iterations (int) â Maximum number of iterations through the corpus when inferring the topic
    distribution of a corpus.
    
    
    eval_every (int, optional) â Log perplexity is estimated every that many updates. 
    Setting this to one slows down training by ~2x
    
    Returns 
    
    lst_corpus list contaning 
        model name - model_(Serial No = no. of corpus provided in the corpus list)_top_(no.of topic taken for modeling 
        as per list provided). Ex. model_3_top_10 
        model - the lda model
        corpus - Initial corpus provided
        dictionary - initially provided
        
        Such a list is created, since it is helpful for executing pyLDAviz
    '''
    
    
    c=1
    lst_corpus=[]
    for dict_corp_tup in corpus_lst:
        lst=[]
        for topic_sz in topic_lst:
            corpus=tfidf_vec(dict_corp_tup[1])
            model_name='model'+'_'+str(c)+'_'+'top'+'_'+str(topic_sz)
            model= LdaModel(corpus=corpus, num_topics=topic_sz,id2word=dict_corp_tup[0],passes=passes,
                                                                 iterations=iterations,chunksize=chunksize,
                                                                 eval_every=eval_every)
            
            lst.append(('model'+'_'+str(c)+'_'+'top'+'_'+str(topic_sz),model,dict_corp_tup[1],dict_corp_tup[0]))
        c+=1
        lst_corpus.append(lst)
    return lst_corpus  
```

```python
# Set training parameters.
num_topics = [3,6,10]
chunksize = 500
passes = 30
iterations = 400
eval_every = 2  
```

```python
model_lst= lda_app(corp_dict_lst,num_topics,passes,chunksize,iterations,eval_every)
```

```python
def topic_print(model,no_of_words):
    topics = model.print_topics(num_words=no_of_words)
    for topic in topics:
        print(topic)  
```
